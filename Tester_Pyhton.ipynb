```python
import csv

lineasBuenasCSV = []

with open('nuestroCSV.csv', newline='') as csvfile: #Si no se pone nada, el fichero se abre en modo lectura
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    #delimiter = separador #quotechar = ignorame lo que está dentro
    next(spamreader) #para saltar la primera linea
    for linea in spamreader:
        #print(', '.join(row))
        print(linea)

        genero = linea [0].lower().split(' ')
        reproducciones = linea[1]
        pegada = 0
        cancion = linea [-1].upper()

        if(genero[0]=='rock'):
          genero = 1
        elif(genero[0]=='classic'):
          genero = 2
        elif(genero[0]=='urban'):
          genero = 3
        else:
          genero = -1

        #reproducciones
        if(reproducciones[-1]=='B' or reproducciones[-1]=='M'):
          if(reproducciones[-1]=='B'):
            reproducciones = float(reproducciones[:-1]) * 10**9
          else:
            reproducciones = float(reproducciones[:-1]) * 10**6
        else:
          reproducciones = float(reproducciones)


        #Pegada
        pegada = float(linea[2]) * float(linea[3]) * float(linea[4]) * float(linea[5]) / 5
        pegada = round(pegada,2)

        lineaDestino = [genero, reproducciones, pegada, cancion]
        print(lineaDestino)
        lineasBuenasCSV.append(lineasBuenasCSV) #Añadimos a linea transformada

print(lineasBuenasCSV)


for linea in lineasBuenasCSV:
  print(lineasBuenasCSV)
```
